```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'F:\Data_analytics_internship_task\assi_2.csv',encoding='utf-8',parse_dates=True).set_index('SKU')
data.head()
```

```python
data.shape
```

```python
data_date_by=data.T
data_date_by.reset_index(inplace=True)
print(data_date_by.columns)
data_date_by
```

```python
data_date_by.columns
```

```python
month=[]
for i in data_date_by['index']:
    month.append(i[3:])
data_date_by['month']=month
data_date_by
```

```python
#What is the monthly & quaterly sales data by SKU
print(f"The monthly sales data by each SKU\n{'--'*40}")
monthly_sales=data_date_by.groupby(['month']).sum()
monthly_sales
```

```python
#Which are the top 3 months in terms of overall sales
sum_by_month=monthly_sales.sum(axis=1,)
monthly_sales['sum_of_sales']=sum_by_month

top_3=monthly_sales.nlargest(3,'sum_of_sales')
print("The top 3 months based on the overall sales")
print('-'*69)
top_3['sum_of_sales']
```

```python
#Which SKUs are outliers in terms sales in any month
scat=monthly_sales.T
scat.reset_index(inplace=True)
scat
```

```python
for x in scat.columns[1:]:
    plt.title(x)
    plt.scatter(x=scat[x],y=scat['SKU'])
    plt.xlabel(x)
    plt.ylabel('SKU')
    plt.show()
```
